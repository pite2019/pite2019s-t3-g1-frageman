```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

dataBoston = load_boston()
df = pd.DataFrame(data = dataBoston.data, columns=dataBoston.feature_names)
df.head()
```

```python
df.hist(column = ['AGE','RM'], bins=30)
reg = LinearRegression()
X = df.loc[:,['AGE','RM']]
y = pd.DataFrame(data = dataBoston.target)
reg.fit(X, y)
```

```python
prediction = reg.predict(X)
plt.scatter(X['AGE'],y)
```

```python
plt.plot(X['AGE'],prediction) 
```

```python
prediction = reg.predict(X)
plt.scatter(X['RM'],y)
```

```python
plt.plot(X['RM'],prediction) 
```
